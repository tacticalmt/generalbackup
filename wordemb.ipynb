```python
def zsl_vec(path):
    
    #if train_data==True:
            #dir_class='trainingset'
            #root_label=os.path.abspath('../dataset/Animals_with_Attributes2')
    #else:
            #dir_class='testingset'
                
        label_path=[x.path for x in os.scandir(path) if 'testclasses' in x.path or 'trainclasses' in x.path]
        print(label_path)
        word=[]
        for i,file_label in enumerate(label_path):
            f=open(file_label,'r')
            while True:
                temp_w=f.readline().strip()
                if not temp_w:
                     break
                word.append(temp_w)
            f.close()
            
        vec_dict={}
        path_dir=[]
        label_list=word
        #print(label_list)
        #vec_se=0
        for key_word in label_list:
            print(key_word)
            if '+' in key_word:
                v=key_word.split('+',1)
                print(key_word)
            else:
                v=[key_word]
                print(v)
            for i,item in enumerate(v):
                if i==0:
                    vec_se=model[item]
                else:
                    vec_se=vec_se+model[item]
                    #vec_se=first_temp
            vec_dict[key_word]=vec_se
        return vec_dict
            
                    
                
            
```

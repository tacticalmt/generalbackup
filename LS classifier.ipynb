```python
import numpy as np
import torch
import torch.nn as nn
import scipy
```

```python
#读数据
data=np.loadtxt("irisdatamodified.data",delimiter=',')
train_data1=data[0:40,0:4] #第一类
train_data2=data[50:90,0:4]#第二类
train_data3=data[100:140,0:4]#3
train_label1=data[0:40,-1:]
train_label2=data[50:90,-1:]
train_label3=data[100:140,-1:]
train_X=np.vstack([train_data1,train_data2,train_data3])
train_Y=np.vstack([train_label1,train_label2,train_label3])
print(train_X,train_Y)
test_data1=data[40:50,0:4]
test_data2=data[90:100,:-1]
test_data3=data[140:150,:-1]
test_label1=data[40:50,-1:]
test_label2=data[90:100,-1:]
test_label3=data[100:140,-1:]
test_X=np.vstack([test_data1,test_data2,test_data3])
test_Y=np.vstack([test_label1,test_label2,test_label3])
print(test_X,test_Y)
```

```python
inv_se=np.linalg.inv(np.dot(train_X.T,train_X))
X_y=np.dot(train_X.T,train_Y)#np.dot(test_X,np.dot(inv_se,np.dot(train_X.T,train_Y)))
w=np.dot(inv_se,X_y)
print(np.dot(test_X,w))
print(np.dot(test_X,np.dot(inv_se,np.dot(train_X.T,train_Y))))
```
